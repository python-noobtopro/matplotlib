## Gas Data in Simple Line Chart / Graph

#### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing from a different folder
# Note: ./ shifts to one level up and then we find the folder and data file in it

gas = pd.read_csv('./matplotlib_tutorial-master/gas_prices.csv')
gas
```

```python
plt.figure(figsize= (10,8))
plt.plot(gas.Year, gas.USA, 'b.-', label='USA')
plt.plot(gas.Year, gas['South Korea'], 'r.-', label='South Korea')
# plt.plot(gas.Year, gas.Germany, 'c.-', label='Germany')
plt.plot(gas.Year, gas.Italy, 'm.-', label='Italy')
# plt.plot(gas.Year, gas.Mexico, 'y.-', label='Mexico')
plt.plot(gas.Year, gas.Japan, 'g.-', label='Japan')
plt.title('Gas Price (per gallon) over Year (in USD)', fontdict={'fontsize':15})
plt.ylabel('US Dollars', fontdict={'fontsize':12})
plt.xlabel('Year', fontdict={'fontsize':12})
plt.xticks(gas.Year[::3])
plt.legend()


plt.show()
```

```python
plt.figure(figsize= (10,8))
for items in gas:
    if items != 'Year':
        plt.plot(gas.Year, gas[items], '.-',label=items)

plt.title('Gas Price (per gallon) over Year (in USD)', fontdict={'fontweight':'bold', 'fontsize':15})
plt.ylabel('US Dollars', fontdict={'fontsize':12})
plt.xlabel('Year', fontdict={'fontsize':12})
plt.xticks(gas.Year[::3])
# plt.legend(loc=2)
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
# To place legend outside the plot

plt.savefig('gasprice.png', bbox_inches = 'tight', facecolor='white', transparent = False, dpi=300)

# Note: In saving as png the problem faced was that only the plot portion was solid, the axes was transparent 
# To remove that problem you can give facecolor in the parameter
# Also, bbox_inches = 'tight' helps us to save the legend completely 
plt.show()

        
```

### Making a Pie Chart around weight og FIFA players

```python
fifa['Count'] = 0
fifa.groupby(['Weight']).count()['Count']
```

#### But we can't make this random distribution in Pie chart. 
#### We have to make categories and for that we have to convert the weight string in integer

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]
fifa.Weight
```

#### Now making the category

```python
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight > 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight > 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight > 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[(fifa.Weight > 200)].count()[0]
```

```python
# Plotting the Pie Chart

weights = [light, light_medium, medium, medium_heavy, heavy]

# There are different styles (color presets) for pie that you can use directly
plt.style.use('ggplot')
plt.figure(figsize=(8,6))
plt.title('Weight Distribution of FIFA Players (in lbs)')
labels = ['Less than 125 lbs', 'Between 125 and 150lbs', 'Between 150 and 175lbs', 'Between 175 and 200 lbs', 'More than 200lbs']

# To make the percentage value clearer we can use pctdistance and explode as parameter
explode=(0.2, 0.1, 0, 0, 0.18)
plt.pie(weights, labels=labels, autopct='%1.2f%%', pctdistance=0.7, explode=explode)

# plt.savefig('fifapieweight.png', bbox_inches = 'tight', facecolor='white', transparent = False, dpi=300)
plt.show()
```

## FIFA Team Comparision (Box and Whisker Plot)

#### Suppose we want to compare the overall strength of different FIFA Clubs

```python
fifa
```

```python
# First of all we need to segregate the overall points of all players of a club

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
manch = fifa.loc[fifa.Club == 'Manchester United']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']
```

```python
# Plotting the Graph
plt.style.use('default')
plt.figure(figsize=(6,7))
plt.title('Comparision of Overall Strength of Soccer Clubs')
labels = ['FC Barcelona', 'Manchester United', 'NE Revolution']
plt.ylabel('Overall Rating')

# patch_artist will allow us to fill color in the box
# other parameter is medianprops, to set the prop of horizontal line in box
boxes = plt.boxplot([barcelona, manch, revs], labels = labels, patch_artist=True, medianprops={'linewidth':1.5, 'color':'yellow'})

# To set individual color of the box edge we can do this
# See the use of zip
colors=['#c42958', '#f52c2c', '#2510b0']
for box, color in zip(boxes['boxes'], colors):
    box.set_facecolor(color)
    
# plt.savefig('fifaboxandwhisker.png', bbox_inches = 'tight', facecolor='white', transparent = False, dpi=300)
plt.show()
```
