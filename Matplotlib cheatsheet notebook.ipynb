```python
import matplotlib.pyplot as plt
import pandas as ps
import numpy as np
```

## Making a basic (x,y) graph

```python
# Suppose

x = [0,1,2,3,4]
y = [0,7,11,3,5]
```

```python
# We want to plot the above on a graph simply

plt.plot(x,y)
```

### Beautify the graph

#### We have various tools to further beautify the above graph

#### 1. Removing the unwanted details [<matplotlib.lines.Line2D at 0x1c635420190>]

```python
plt.plot(x,y)
plt.show()
```

#### 2. Adding a title to the graph and axes

```python
plt.plot(x,y)
plt.title('Our first graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

#### 3. Changing fontname and fontsize

```python
plt.plot(x,y)
plt.title('Our first graph', fontdict={'fontname':'Comic Sans MS', 'fontsize':'20'})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

#### 4. Changing the markers on x and y axis

```python
plt.plot(x,y)
plt.title('Our first graph', fontdict={'fontname':'Comic Sans MS', 'fontsize':'20'})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks([0,1,2,3,4,5])
plt.yticks([0,5,10,15])

plt.show()
```

#### 5. Adding label for line (known as legend)

```python
plt.plot(x,y, label='x = y*2+x')

plt.title('Our first graph', fontdict={'fontname':'Comic Sans MS', 'fontsize':'20'})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks([0,1,2,3,4,5])
plt.yticks([0,5,10,15])

plt.legend()

plt.show()
```

#### 6. Changing the color / width of line

```python
plt.plot(x,y, label='x = y*2+x', color='red', linewidth='4')

plt.title('Our first graph', fontdict={'fontname':'Comic Sans MS', 'fontsize':'20'})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks([0,1,2,3,4,5])
plt.yticks([0,5,10,15])

plt.legend()

plt.show()
```

#### 4. Adding markers for each plotted point on the line

```python
plt.plot(x,y, label='x = y*2+x', color='red', linewidth=2, marker='.', markersize=10, markeredgecolor='b', linestyle='--')

plt.title('Our first graph', fontdict={'fontname':'Comic Sans MS', 'fontsize':'20'})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks([0,1,2,3,4,5])
plt.yticks([0,5,10,15])

plt.legend()

plt.show()
```

### Using shorthand method to plot a Graph

#### fmt = '[color][marker][line]'

```python
plt.plot(x,y, 'b^--', label='x = y*2+x')

plt.title('Our first graph', fontdict={'fontname':'Comic Sans MS', 'fontsize':'20'})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks([0,1,2,3,4,5])
plt.yticks([0,5,10,15])

plt.legend()

plt.show()
```

#### This method is useful when you want to make a line very quickly

 ## Making more complex graph

```python
sq = np.arange(0,5.5,0.5)
sq

# This code gives an array of numbers between 0 to 5.5 with stepsize of 0.5
```

```python
# Suppose we want to plot the square function of this function

plt.plot(sq, sq**2)
plt.show()
```

```python
# Combining with the above graph to do more stuff

plt.plot(x,y, 'b^--', label='x = y*2+x')

plt.title('Our first graph', fontdict={'fontname':'Comic Sans MS', 'fontsize':'20'})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks([0,1,2,3,4,5])
#plt.yticks([0,5,10,15])   # We are using the default ticks to make grapdh clear

plt.plot(sq, sq**2, 'r', label='y = x^2')

plt.legend()

plt.show()
```

#### Suppose we want to make a projection graph

```python
sq
```

```python
plt.plot(sq[:7], sq[:7]**2, 'r', label='y = x^2')
plt.plot(sq[6:], sq[6:]**2, 'r--')
plt.legend()
plt.show()

# This snippet will take the first six values of sq and make it solid and dotted beyond that value
```

## Resizing the Graph and increasing the quality

```python
plt.figure(figsize=(5,4), dpi=100)

# a dpi of 300 is a good dpi


plt.plot(sq[:7], sq[:7]**2, 'r', label='y = x^2')
plt.plot(sq[6:], sq[6:]**2, 'r--')
plt.legend()
plt.show()
```

#### Suppose we want to move out the legend from the graph

```python
plt.figure(figsize=(5,4), dpi=100)

plt.plot(sq[:7], sq[:7]**2, 'r', label='y = x^2')
plt.plot(sq[6:], sq[6:]**2, 'r--')

####
plt.legend(bbox_to_anchor=(1,1), loc="upper left")
####

plt.show()
```

## Saving a graph as a figure

```python
plt.plot(x,y, 'b^--', label='x = y*2+x')

plt.title('Our first graph', fontdict={'fontname':'Comic Sans MS', 'fontsize':'20'})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks([0,1,2,3,4,5])
#plt.yticks([0,5,10,15])   # We are using the default ticks to make grapdh clear

plt.plot(sq, sq**2, 'r', label='y = x^2')

plt.legend()

plt.savefig('mygraph.png', dpi=300)

# Always remeber to save the graph before show() otherwise it will save a blank image

plt.show()
```

#### Note: The problem with simple 'png' save is it saves the plot in solid and axes transparent. We can use certain parameters to avoid that as follows:

```python
# plt.savefig('mygraph.png', bbox_inches = 'tight', facecolor='white', transparent = False, dpi=300)
```

## Bar Chart

### Making a simple Bar Chart

```python
a = ['A', 'B', 'C']
b = [153,52,97]

bar = plt.bar(a,b)
```

### Making the bar chart more visual

```python
a = ['A', 'B', 'C']
b = [153,52,97]

bar = plt.bar(a,b)
bar[0].set_hatch('/')
bar[1].set_hatch('O')
bar[2].set_hatch('*')
```

```python
# We can also do the above thing as

a = ['A', 'B', 'C']
b = [153,52,97]

bar = plt.bar(a,b)

patterns = ['/','O','*']
# bar[0].set_hatch('/')
# bar[1].set_hatch('O')
# bar[2].set_hatch('*')

for i in bar:
    i.set_hatch(patterns.pop(0))

```

## Making Pie chart

```python
a = [10,20,30,40]
labels = ['A', 'B', 'C', 'D']

plt.pie(a, labels=labels)
plt.show()
```

#### Suppose we want to explode 'A'

```python
explode = (0, 0.09, 0, 0)

plt.pie(a, labels=labels, explode=explode, autopct='%1.0f%%')   # Also adding the percentage
plt.show()
```
